```python
import pandas as pd
import import_ipynb
from DataCleaning import clean_column, semi_clean_column, spaces
```

```python
file_name = '../Uncleaned Data/all-the-news/articles{}.csv'
df1 =  pd.read_csv(file_name.format(1), header=0,names=['title','publication', 'text'], usecols=[2,3,9])
df2 =  pd.read_csv(file_name.format(2), header=0,names=['title','publication','text'], usecols=[2,3,9])
df3 =  pd.read_csv(file_name.format(3), header=0,names=['title','publication','text'], usecols=[2,3,9])
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
frames = [df1, df2, df3]
result = pd.concat(frames)
result.reset_index(inplace=True, drop=True)
result.shape
```

```python
result.head()
```

```python
result.tail()
```

```python
result.groupby('publication').agg('count')
```

```python
keep = ["Reuters", "Business Insider", "Atlantic"]
```

```python
result = result[result['publication'].isin(keep)]
result.reset_index(inplace=True, drop=True)
print(result.shape)
result.groupby('publication').agg('count')
```

```python
result['label'] = 1
result.drop(['publication'],axis=1,inplace=True)
result.head()
```

```python
space_text = []
for i in range(len(result['text'])):
    try:
        if result['text'].iloc[i] in spaces:
            space_text.append(i)
        elif not result['text'].iloc[1]:
            print("BRUH WTF")
    except TypeError:
        print("TYPE ERROR")
        print(i)
print(space_text)
result['text'].iloc[8104]
```

```python
result['title'] = result['title'].apply(lambda x: str(x))
result['text'] = result['text'].apply(lambda x: str(x))
for i in spaces:
    result = result[(result['text'] != i)]
result = result[(result['title'] != 'nan') & 
                                  (result['text'] != 'nan') &
                                  (result['title'] != '')]
result_clean = semi_clean_column(result, 'text', 'clean_text')
result_clean = semi_clean_column(result_clean, 'title', 'clean_title')
result_clean.head()
# result['text'] = result['text'].apply(lambda x: str(x))
# result = result[(result['text'] != 'nan') & (result['text'] != '')]
# print(result.shape)
# result_clean = clean_column(result, 'text', 'clean_text')
# print(result_clean.shape)
# result_clean.to_csv('../Cleaned Data/Kaggle3_Real.csv')
```

```python
# result_clean.drop(['title', 'text'],inplace=True,axis=1)
result_clean = result_clean[['clean_text','clean_title','label']]
result_clean.head()
```

```python
for i in spaces:
    result_clean = result_clean[(result_clean['clean_text'] != i)]
```

```python
space_text = []
for i in range(len(result_clean['clean_text'])):
    try:
        if result_clean['clean_text'].iloc[i] in spaces:
            space_text.append(i)
        elif not result_clean['clean_text'].iloc[1]:
            print("BRUH WTF")
    except TypeError:
        print("TYPE ERROR")
        print(i)
print(space_text)
```
