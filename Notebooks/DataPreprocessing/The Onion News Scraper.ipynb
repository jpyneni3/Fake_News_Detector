```python
from psaw import PushshiftAPI
from requests_html import HTMLSession
from urllib.parse import urlparse
import pandas as pd

session = HTMLSession()

api = PushshiftAPI()
results = list(api.search_submissions(subreddit='TheOnion', filter=['url', 'title'], limit=10000))

onion_links = set()
data = []
counter = 0

for r in results:
    source_components = urlparse(r.url).netloc.split('.')
    source = ".".join(source_components[1:])
    if source == 'theonion.com' and r.url not in onion_links:
        onion_links.add(r.url)
        s = session.get(r.url)
        content = s.html.find('.sc-77igqf-0')
        
        if content:
            data.append([r.url, r.title, content[0].text])
            counter += 1
            if (counter % 500 == 0):
                print('Scraped ' + str(counter) + ' articles')

df = pd.DataFrame(data, columns = ['url', 'title', 'text'])

df.to_csv('TheOnion.csv')
print(df.shape)
df.head(100)
```
